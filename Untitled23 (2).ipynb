```python
heatmap_data = df.groupby('Product Category')['Units Sold'].sum().to_frame()

matrix = pd.crosstab(df['Product Category'], df['Region'])

plt.figure(figsize=(8,6))
sns.heatmap(matrix, annot=True, cmap='YlGnBu')
plt.title('Product Category vs Region')
plt.show()
```

The heatmap shows the distribution of product categories across regions. Each product category contains 40 transactions. Beauty Products and Home Appliances are associated with Europe, Books and Electronics with North America, and Clothing and Sports with Asia. Since all categories have the same number of transactions, the heatmap reflects the dataset structure rather than differences in sales performance.

```python
product_summary = df.groupby('Product Name').agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Count=('Product Name', 'count')
).reset_index().sort_values(
    by='Count',
    ascending=False
)

print(product_summary.head(10))
```

Only a few products appeared in multiple transactions, with Garmin Forerunner 945 generating the highest revenue (1,599.97). Most products were purchased only once, reflecting a diverse range of products and customer preferences.

```python
payment_count = df['Payment Method'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    payment_count,
    labels=payment_count.index,
    autopct='%1.1f%%'
)

plt.title('Payment Method Distribution')
plt.show()
```

Credit Card was the most frequently used payment method (50%), followed by PayPal (33.3%) and Debit Card (16.7%).
